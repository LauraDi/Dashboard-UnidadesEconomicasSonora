Importamos librerias necesarias

```python
import pandas as pd
from pandas_profiling import ProfileReport 
import sweetviz
```

Realizaremos una funcion para ejecutar un EDA automatico, ya que como tenemos varios conjuntos de datos, será conveniente tener una funcion para solo llamarla y no tener que producir mas código.

```python
def edaAutomatico(direccion, nombre):
    insurance = pd.read_csv(direccion)
    profile = ProfileReport(
        insurance, 
        explorative=True,
        title='Reporte de insurance.csv', 
        html={'style':{'full_width':True}}
    ) 

    profile.to_file(nombre + "- profiler.html")
```

Enviaremos la direccion del conjunto de datos asi como el nombre para identificar al archivo html que nos generará profiler.

```python
edaAutomatico("../DatosUnidadesEconomicas/Sector-Municipality-Year.csv", "UnidadesEconomicas")
edaAutomatico("../DatosCensos/2000.csv", "Censo2000")
edaAutomatico("../DatosCensos/2010.csv", "Censo2010")
edaAutomatico("../DatosCensos/2020.csv", "Censo2020")
edaAutomatico("../datosIncidentes/2017.csv", "Incidentes2017")
```
