### Descargando datos del año 2010

Visualizando los datos

```python
nombreArchivo= "./DatosCensos/2000.csv"
df2000 = pd.read_csv(nombreArchivo)
df2000.head()
```

Como solo queremos el total de poblacion por municipio, filtraremos todos los renglones que en la columna "nom_loc" tengan el texto "Total del Municipio"

```python
df2000 = df2000[df2000.nom_loc == "TOTAL MUNICIPAL"]
```

Ahora, nos quedaremos solamente con la columna que queremos --> El nombre del municipio y su respectiva poblacion total

```python
df2000 = pd.DataFrame(df2000, columns=["nom_mun", "pobtot"])
df2000.reset_index(drop=True, inplace=True) #reseteamos index 
df2000.rename({'nom_mun': 'municipio', 'pobtot': '2000'}, axis=1, inplace=True) #cambiamos de nombre las columnas
df2000
```

Realizaremos el mismo proceso con los archivos del censo del 2010 y 2020

- 2010

```python
nombreArchivo2= "./DatosCensos/2010.csv"
df2010 = pd.read_csv(nombreArchivo2)
df2010 = df2010[df2010.nom_loc == "Total del Municipio"]
df2010 = pd.DataFrame(df2010, columns=["nom_mun", "pobtot"])
df2010.reset_index(drop=True, inplace=True) #reseteamos index 
df2010.rename({'nom_mun': 'municipio', 'pobtot': '2010'}, axis=1, inplace=True) #cambiamos de nombre las columnas
df2010
```

- 2020

```python
nombreArchivo3= "./DatosCensos/2020.csv"
df2020 = pd.read_csv(nombreArchivo3)
df2020 = df2020[df2020.NOM_LOC == "Total del Municipio"]
df2020 = pd.DataFrame(df2020, columns=["NOM_MUN", "POBTOT"])
df2020.reset_index(drop=True, inplace=True) #reseteamos index 
df2020.rename({'NOM_MUN': 'municipio', 'POBTOT': '2020'}, axis=1, inplace=True) #cambiamos de nombre las columnas
df2020
```

Combinacion de los datafarames

```python
CensoSonora = pd.merge(df2000, df2010)
CensoSonora = pd.merge(CensoSonora, df2020)
CensoSonora
```

Guardamos el datframe en un archivo parquet

```python
CensoSonora.to_parquet('SonoraCenso2000-2020.gzip')
```

Veamos el tipo de cada columna

```python
CensoSonora.dtypes
```

Veamos si tenemos datos nulos

```python
CensoSonora.isnull().sum()
```
